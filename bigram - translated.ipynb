```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import pandas as pd
from nltk import sent_tokenize, word_tokenize

from gensim.models import Phrases
from gensim.models.phrases import Phraser
from gensim import models
import gensim
```

```python
# modelo para tradução
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-ROMANCE")
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ROMANCE", model_max_length=150)
```

```python
# leitura de um arquivo de texto
sent_lst = []
with open("titulos.txt","r") as fd:
    for line in fd:
        sent_lst.append(">>pt<< " + line)
```

```python
sent_lst0 = sent_lst[:100]
sent_lst1 = sent_lst[100:200]
sent_lst2 = sent_lst[200:300]
sent_lst3 = sent_lst[300:]
```

```python
# tokenizando as sentenças
encoded0 = tokenizer(sent_lst0, return_tensors='pt', padding=True)
```

```python
encoded1 = tokenizer(sent_lst1, return_tensors='pt', padding=True)
```

```python
encoded2 = tokenizer(sent_lst2, return_tensors='pt', padding=True)
```

```python
encoded3 = tokenizer(sent_lst3, return_tensors='pt', padding=True)
```

```python
# traduzindo
sent_translation0 = model.generate(**encoded0)
```

```python
# traduzindo
sent_translation1 = model.generate(**encoded1)
```

```python
# traduzindo
sent_translation2 = model.generate(**encoded2)
```

```python
# traduzindo
sent_translation3 = model.generate(**encoded3)
```

```python
# preparando a saída
output0 = tokenizer.batch_decode(sent_translation0, skip_special_tokens=True)
tok_lst0 = [ word_tokenize(sent) for sent in output0]
```

```python
# preparando a saída
output1 = tokenizer.batch_decode(sent_translation1, skip_special_tokens=True)
tok_lst1 = [ word_tokenize(sent) for sent in output1]
```

```python
# preparando a saída
output2 = tokenizer.batch_decode(sent_translation2, skip_special_tokens=True)
tok_lst2 = [ word_tokenize(sent) for sent in output2]
```

```python
# preparando a saída
output3 = tokenizer.batch_decode(sent_translation3, skip_special_tokens=True)
tok_lst3 = [ word_tokenize(sent) for sent in output3]
```

```python
tok_lst = tok_lst0 + tok_lst1 + tok_lst2 + tok_lst3
```

```python
bigram = Phrases(tok_lst, min_count=1, threshold=2,delimiter=' ')
bigram_phraser = Phraser(bigram)
```

```python
bigram_token = []
for sent in tok_lst:
    bigram_token.append(bigram_phraser[sent])
```

```python
#now you can make dictionary of bigram token 
dict_ = gensim.corpora.Dictionary(bigram_token)
```

```python
# id_token -> frequency mapping in the dict_ object
dict_.cfs
```

```python
hist = []
for idx, tok in dict_.iteritems():
    hist.append((tok, dict_.cfs[idx]))
df_hist = pd.DataFrame(hist,columns = ["token","frequency"])
```
